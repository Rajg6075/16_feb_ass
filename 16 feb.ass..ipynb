```python
Q1. What is a database? Differentiate between SQL and NoSQL databases.

ANS= A database is an organized collection of structured information, or data, typically stored electronically in a computer system.

SQL Databases:

Relational data model (tables with rows and columns).
Rigid schema structure.
Standardized SQL query language.
Strong consistency, ACID properties.
Suited for structured data, complex querying, transactions.
Vertical scalability.


NoSQL Databases:

Various data models (document, key-value, etc.).
Flexible or schema-less structure.
Model-specific query languages.
Emphasis on scalability and availability.
Suited for unstructured/semi-structured data, rapid scalability.
Horizontal scalability.


Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

ANS= DDL (Data Definition Language) is a subset of SQL that deals with defining and managing the structure of databases, tables, and other database objects. It includes commands like CREATE, DROP, ALTER, and TRUNCATE.
Let's break down each command with examples:

CREATE is used to make new database objects (like tables).
DROP is used to delete database objects.
ALTER is used to modify existing objects' structures.
TRUNCATE is used to remove all data from a table.

Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

ANS= DML (Data Manipulation Language) is a subset of SQL that deals with manipulating data stored in a database. It includes commands like INSERT, UPDATE, and DELETE.

INSERT adds new data to a table.
UPDATE modifies existing data in a table.
DELETE removes data from a table.

Q4. What is DQL? Explain SELECT with an example.

ANS= DQL (Data Query Language) is a subset of SQL that focuses on retrieving data from a database. The primary DQL command is SELECT, which is used to query and retrieve data from one or more tables.

SELECT * retrieves all columns from the specified table.
SELECT ProductName, Price retrieves only the "ProductName" and "Price" columns.
WHERE clause is used to filter rows based on specific conditions.
ORDER BY sorts the retrieved rows based on the specified column(s).
DESC sorts in descending order.

Q5. Explain Primary Key and Foreign Key.

ANS= Primary Key:
A primary key is a column or set of columns in a database table that uniquely identifies each record (row) in that table. It ensures that each record has a unique identifier, which helps in data integrity and efficient data retrieval.

Foreign Key:
A foreign key is a column or set of columns in a table that establishes a link between the data in two different tables. It creates a relationship that enforces referential integrity between the two tables. The foreign key column in one table refers to the primary key column in another table.

Q6. Write a python code to connect MySQL to python. Explain the cursor() and execute() method.

ANS=  here's an example of how to connect to MySQL using Python and then I'll explain the cursor() and execute() methods:
    
    import mysql.connector

# Connect to MySQL
mydb = mysql.connector.connect(
    host="localhost",
    user="username",
    password="password",
    database="mydatabase"
)

# Create a cursor object
cursor = mydb.cursor()

# Execute an SQL query
query = "SELECT * FROM Customers"
cursor.execute(query)

# Fetch and print the results
results = cursor.fetchall()
for row in results:
    print(row)

# Close the cursor and connection
cursor.close()
mydb.close()


Q7. Give the order of execution of SQL clauses in an SQL query.

ANS= In a standard SQL query, the order of execution of clauses follows a general sequence:

SELECT: This clause specifies the columns you want to retrieve data from. It's typically the first clause in a query and determines the data you're interested in.

FROM: This clause indicates the table(s) from which you're retrieving data. It defines the source of the data for your query.

JOIN: If your query involves multiple tables and you're combining data from them, JOIN clauses are used to specify how the tables are related. JOINs help in combining data from different tables based on common columns.

WHERE: The WHERE clause filters the rows that meet specific conditions. It narrows down the data you want to retrieve based on specified criteria.

GROUP BY: If you're performing aggregation operations (like SUM, COUNT, AVG) on grouped data, you use the GROUP BY clause to group rows based on one or more columns.

HAVING: The HAVING clause filters aggregated data after GROUP BY operations. It's similar to the WHERE clause but works with aggregated values.

ORDER BY: The ORDER BY clause sorts the result set based on one or more columns. It allows you to specify the order (ascending or descending) in which the data should be presented.

LIMIT/OFFSET (Optional): Some databases support the LIMIT clause, which restricts the number of rows returned. The OFFSET clause (often used with LIMIT) is used to skip a specified number of rows before starting to return data.
```
